```python
import pandas as pd
df=pd.read_csv('music1.csv', encoding='ISO-8859-1')
df
```

```python
df.drop('duration',axis=1,inplace=True)
df.drop('popularity',axis=1,inplace=True)
df
```

```python
df['combined_features']=df['Genre']+df['language']
df['combined_features']
```

```python
ds  = df.drop(columns=['Genre','language','poster'])
ds
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
c=cv.fit_transform(df['combined_features'].values.astype('U')).toarray()

```

```python
#Genre_series=pd.Series(['Genre'])
#vector=cv.fit_transform(['Genre'].())
#vector.shape
c.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c)
similarity
```

```python
# ds=df['song_name']+df['singer']+df['Genre']+df['language']
# ds
idx=ds[ds['song_name']=="Sharbat Wangu Ghut Bhar Lan"].index[0]
idx
```

```python
# #idx=df[df['song_name']=="Sharbat Wangu Ghut Bhar Lan"].index[0]
# def recommand(song_name):
#   index=ds[ds['song_name']==song_name].index[0]
import random
distance = sorted(list(enumerate(similarity[idx])), reverse=True, key=lambda vector:vector[1])
# for i in distance[0:5]:
s=[ds.iloc[i[0]].song_name for i in distance[1:10]]
random.shuffle(s)
for i in s[:5]:
    print(i)
```

```python
def recommand(song_name):
    index=ds[ds['song_name']==song_name].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
    s1=[ds.iloc[i[0]].song_name for i in distance[1:10]]
    random.shuffle(s1)
    for i in s1[0:5]:
        print(i)
```

```python
recommand("Aaja Sonie")
```
